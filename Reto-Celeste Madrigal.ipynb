```python
#Importar librerias
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from datetime import datetime
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
# Paso 2: Leer el archivo CSV
EmpleadosAttrition = pd.read_csv('empleadosRETO.csv', delimiter=';')
EmpleadosAttrition
```

```python
# Paso 3: Eliminar columnas irrelevantes
columnas_a_eliminar = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']

# Elimina las columnas
EmpleadosAttrition = EmpleadosAttrition.drop(columnas_a_eliminar, axis=1)

# Muestra el DataFrame resultante
print(EmpleadosAttrition)
```

```python
#
# Convierte 'HiringDate' a formato de fecha
EmpleadosAttrition['HiringDate'] = pd.to_datetime(EmpleadosAttrition['HiringDate'], errors='coerce')

# Paso 5: Crea la columna 'Year' con el año de contratación
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].dt.year.astype('Int64')  # Convierte a entero

#Paso 6: Calcula los años que el empleado lleva en la compañía hasta el año 2018
fecha_actual = datetime(2018, 1, 1)  # Fecha actual considerada para el cálculo
EmpleadosAttrition['YearsAtCompany'] = (fecha_actual - EmpleadosAttrition['HiringDate']).dt.days // 365

# Muestra el DataFrame resultante
print(EmpleadosAttrition[[ 'HiringDate', 'Year', 'YearsAtCompany']])
```

```python
# Paso 8:Renombra la variable 'DistanceFromHome' a 'DistanceFromHome_km'
EmpleadosAttrition.rename(columns={'DistanceFromHome': 'DistanceFromHome_km'}, inplace=True)

# Paso 7 y 9:Elimina las letras "km" y convierte la columna a tipo numérico
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome'].astype(str).str.rstrip('km')
EmpleadosAttrition['DistanceFromHome'] = pd.to_numeric(EmpleadosAttrition['DistanceFromHome'], errors='coerce')

# Muestra el DataFrame resultante
print(EmpleadosAttrition[['DistanceFromHome_km', 'DistanceFromHome']])
```

```python
# Paso 10:Borra las columnas especificadas
EmpleadosAttrition = EmpleadosAttrition.drop(['Year', 'HiringDate', 'DistanceFromHome_km'], axis=1)

# Muestra el DataFrame resultante
print(EmpleadosAttrition.head())
```

```python
# Paso 11: Genera el DataFrame con el sueldo promedio por departamento
SueldoPromedioDepto = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().reset_index()

# Asigna el resultado a la variable SueldoPromedio
SueldoPromedio = SueldoPromedioDepto.copy()

# Muestra el DataFrame resultante
print(SueldoPromedioDepto)
```

```python
# Paso 12: Escala la variable MonthlyIncome entre 0 y 1
scaler = MinMaxScaler()
EmpleadosAttrition['MonthlyIncome_scaled'] = scaler.fit_transform(EmpleadosAttrition[['MonthlyIncome']])

# Muestra el DataFrame resultante con la nueva columna escalada
print(EmpleadosAttrition[['MonthlyIncome', 'MonthlyIncome_scaled']])
```

```python
# Paso 13: Convierte variables categóricas a numéricas usando one-hot encoding
EmpleadosAttrition = pd.get_dummies(EmpleadosAttrition, columns=['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Attrition'])

# Muestra el DataFrame resultante con las nuevas columnas numéricas
print(EmpleadosAttrition.head())
```
